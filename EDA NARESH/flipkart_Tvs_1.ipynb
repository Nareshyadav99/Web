```python
%%time
title=[]
price=[]
rating=[]
reviwe=[]
os=[]
off=[]
feature=[]

pages=41
print("pro:" ,end="")

for i in range(1,pages):
    url=(f"https://www.flipkart.com/search?q=smart+tv&page={i}")
    
    response = requests.get(url,headers={'Content-Type': 'text/html; charset=UTF-8','User-Agent': 'Chrome/101.0.0.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0','Accept-Encoding': 'gzip, deflate, br'})
    
    
    html_code=response.text
    #print(html_code)
    
    soup=BeautifulSoup(html_code)
    
    for site in soup.find_all("div",attrs={"class":"tUxRFH"}):
        
        product=site.find("div",attrs={"class":"KzDlHZ"})
        if product is None:
            title.append(np.NAN)
        else:
            title.append(product.text)
            
        prices=site.find('div', attrs={'class' : 'Nx9bqj _4b5DiR'})
        if prices is None:
            price.append(np.NAN)
        else:
            price.append(prices.text)
            
        ratings=site.find("div",attrs={"class":"XQDdHH"})
        if ratings is None:
            rating.append(np.NAN)
        else:
            rating.append(ratings.text)
            
        reviwes=site.find("span",attrs={"class":"Wphh3N"})
        if reviwes is None:
            reviwe.append(np.NAN)
        else:
            reviwe.append(reviwes.text)
            
        operating_sys=site.find("li",attrs={"class":"J+igdf"})
        if  operating_sys is None:
            os.append(np.NAN)
        else:
            os.append( operating_sys.text)
            
        resolutions=site.find("li",attrs={"class":"J+igdf"})
        
        offer=site.find("div",attrs={"class":"yRaY8j ZYYwLA"})
        if offer is None:
            off.append(np.NAN)
        else:
            off.append(offer.text)
            
        features=site.find('ul', attrs={'class' : 'G4BRas'})
        if features  is None:
            feature.append(np.NAN)
        else:
            feature.append(features.text)
    print(">",end="")
print()
```
