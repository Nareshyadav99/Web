# EXPLORATORY DATA ANALYSIS

**Defination**: It stands for Exploratory Data Analysis.It is an approach to analysing and visuliasing datasets to summarise their main characteristics with the heklp of statistical graphs or other visualisation methods

**Purpose**:The main purpose of EDA is to help look at data before making any assumptions. It can help identify obvious errors, as well as better understand patterns within the data, detect outliers or anomalous events, find interesting relations among the variables.
* EDA helps determine how best to manipulate data sources to get the answers you need, making it easier for data scientists to discover patterns, spot anomalies, test a hypothesis, or check assumptions.

## Univariant Analysis

**Defination**: Univariant Analysis involves the examination and interpretation of a single variable in isolation.it focus on understanding the distribution central tendency and a dispersion of a single variable

**Purpose**:univariant analysis is useful for describing and summarizing the characteristics of a individual  variables .it provide insights into the basic properties of the data without considering relationships with other variables.

# Multivariant Analysis 


```python
columns = ['offer_price','product_rating','no_of_ratings','no_of_reviews']#multi variantt
subset_dfx = dfx.loc[:,columns]
sns.pairplot(subset_dfx, hue='product_rating', diag_kind='kde')
plt.suptitle('Pair Plot Colored by Rating', y=1.02)
plt.show()
```

```python
# Grouped bar chart for price comparison across brands
plt.figure(figsize=(20, 10))

# Grouped bar chart for Offer_Price
plt.bar(dfx['Brand'] + '_Offer', dfx['offer_price'], color='skyblue', label='offer_price')

# Grouped bar chart for Original_Price
plt.bar(dfx['Brand'] + '_Original', dfx['original_price'], color='orange', label='original_price')

# Grouped bar chart for Discount_Price
plt.bar(dfx['Brand'] + '_Discount', dfx['discount_price'], color='green', label='discount_price')

plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Price Comparison Across Brands')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Box plot for price comparison across brands
plt.figure(figsize=(10, 12))

# Box plot for Offer_Price
sns.boxplot(x='Brand', y='offer_price', data=dfx, color='skyblue', width=0.25)

# Box plot for Original_Price
sns.boxplot(x='Brand', y='original_price', data=dfx, color='orange', width=0.25)

# Box plot for Discount_Price
sns.boxplot(x='Brand', y='discount_price', data=dfx, color='green', width=0.25)

plt.xlabel('Brand')
plt.ylabel('Price')
plt.title('Price Comparison Across Brands')
plt.xticks(rotation=45)

# Manually adding legend
plt.legend(['Offer Price', 'Original Price', 'Discount Price'])

plt.show()
```
